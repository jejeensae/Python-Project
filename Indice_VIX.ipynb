```python
import urllib 
import bs4
import pandas
from urllib import request

```

```python

url_VIX_index = "https://www.boursorama.com/bourse/indices/cours/historique/1cVIX"
request_text=request.urlopen(url_VIX_index).read()
page = bs4.BeautifulSoup(request_text, "lxml")
```

```python
## On identifie le tableau avec l'indice VIX 

tableau_VIX_index = page.find('table', {'class' : 'c-table'})
table_body = tableau_VIX_index.find('tbody')

### On recherche toutes les lignes du tableau 

rows = table_body.find_all('tr')
##print(rows[1])

## On nettoie les sorties pour 1 lignes
for row in rows:
    cols = row.find_all('td')
    cols = [ele.text.strip() for ele in cols]
    print(cols)
```

```python
##On met tout dans en dictionnaire pour le transformer en dataframe pandas

dico_VIX_index = dict()
for row in rows:
    cols = row.find_all('td')
    cols = [ele.text.strip() for ele in cols]
    if len(cols) > 0 : 
        dico_VIX_index[cols[0]] = cols[1:]
dico_VIX_index
```

```python
## Transformation en DataFrame
data_VIX_index = pandas.DataFrame.from_dict(dico_VIX_index,orient='index')
data_VIX_index
```

```python
## On récupère les têtes du tableaux
for row in rows:
    cols = row.find_all('th')
    print(cols)
    if len(cols) > 0 : 
        cols = [ele.get_text(separator=' ').strip().title() for ele in cols]
        columns_VIX_index = cols
```
